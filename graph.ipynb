```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csvfile = open('./TEST_FILES/copy.csv','r')

data = csvfile.read()
```

```python
data = data.split('\n')
Rows = list()
for row in data:
    row = row.split(';')
    for elem in row: elem = int(elem)
    Rows.append(row)
```

```python
df = pd.DataFrame(Rows, columns=['Size','Copy','Compare'])
dfs = list()
for i in range(0,len(df),14):
    j = i
    dfs.append(df[j:j+14])
```

```python
print(dfs[1])
```

```python

sort = ['Comd sort','Quicksort','Insertion sort']
j = 0

for i in range(0,len(dfs),3):
    sub_dfs = dfs[i:i+3]
    lable = sort[j]
    
    for dataframe in sub_dfs:
        fig1, ax1 = plt.subplots()
        ax1.plot(dataframe['Size'], dataframe['Copy'], '-r', label='Copy')
        ax1.plot(dataframe['Size'], dataframe['Compare'], '-b', label='Compare')
        leg = ax1.legend()
        ax1.set_title(lable)
        
    j+=1
```
